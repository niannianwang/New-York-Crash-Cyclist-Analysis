#### Exploratory Data Analysis
- Plot incidents on a map

```python
fig = px.choropleth(locations=["New York City"], locationmode="USA-states", color=[3], scope="usa")

# Update layout to focus on New York City
fig.update_geos(center=dict(lon=-74.0060, lat=40.7128),  # New York City coordinates
                projection_scale=100,  # Adjust zoom level as needed
                visible=False
               )

# Show the plot
fig.show()
```
